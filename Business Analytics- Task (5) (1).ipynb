# Importing Relevant Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
store=pd.read_csv('SampleSuperstore.csv')
```

```python
data=store[['Ship Mode','Sales', 'Quantity','Discount','Profit']]
```

Weak area which needs to be improved.

#  By using Data Visualization

```python
sns.pairplot(data, hue='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit',data=store,col='Segment')
```

```python
plt.figure(figsize=(20,20))
sns.catplot(x='Ship Mode',y= 'Profit',data=store, kind='bar')
plt.xlabel('Shipping Mode', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(fontsize=10)
```

```python
sns.catplot(x='Category', y='Discount',data=store,kind='box', palette='rainbow')
```

```python
store.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
sns.countplot(data['Ship Mode'],palette='rainbow')
```

```python
sns.countplot(store['Category'],palette='rainbow')
```

```python
sns.countplot(store['Segment'], palette='rainbow')
```

```python
corelation=store.corr()
corelation
```

```python
sns.heatmap(corelation,annot=True,cmap='coolwarm',linecolor='black')
```

```python
plt.figure(figsize=(10,5))
store.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(store['Sub-Category'], hue='Region',data=store)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,12))
sns.countplot(x=store['Ship Mode'],hue= store['Segment'], palette='rainbow')
plt.xlabel('Shipping Mode', fontsize=20)
plt.legend(loc=5, fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,12))
sns.countplot(x=store['State'],palette='coolwarm',order=(store['State'].value_counts().head(5)).index)
plt.xlabel('State',fontsize=20)
plt.ylabel('Ordered by state',fontsize=20)
plt.show()
```

```python
Top_5=store.groupby('State')['Quantity'].sum().reset_index().sort_values(by='Quantity',ascending=True)
```

```python
plt.figure(figsize=(20,8))
sns.catplot(x='State', y='Quantity',data=Top_5.head(), kind='bar', palette='rainbow')
plt.xlabel('State', fontsize=20)
plt.ylabel('Quantity',fontsize=20)
plt.xticks(rotation=90,fontsize=10)
plt.title('Top 5 state')
plt.show()
```
